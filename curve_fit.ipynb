```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);

from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);

print("Model slope:   ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

# Membuat fitur polinomial derajat 2
polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x[:, np.newaxis])

# Membuat model regresi linear
model = LinearRegression()
model.fit(x_poly, y)

# Membuat prediksi
x_fit = np.linspace(0, 10, 1000)
x_fit_poly = polynomial_features.transform(x_fit[:, np.newaxis])
y_fit = model.predict(x_fit_poly)

# Plot hasil
plt.scatter(x, y)
plt.plot(x_fit, y_fit, color='red')
plt.show()
```

```python
# Ganti derajat polinomial menjadi 3
polynomial_features= PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(x[:, np.newaxis])
model.fit(x_poly, y)
x_fit_poly = polynomial_features.transform(x_fit[:, np.newaxis])
y_fit = model.predict(x_fit_poly)

# Plot hasil
plt.scatter(x, y)
plt.plot(x_fit, y_fit, color='green')
plt.show()
```

```python
from sklearn.linear_model import Ridge

# Set parameter alpha untuk regularisasi L2
model = Ridge(alpha=1.0)
model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);

print("Model slope:   ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
from sklearn.linear_model import Lasso

# Set parameter alpha untuk regularisasi L1
model = Lasso(alpha=1.0)
model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);

print("Model slope:   ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
from sklearn.svm import SVR

# Set parameter kernel untuk fungsi kernel linear
model = SVR(kernel='linear')
model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);

print("Model support vectors:", model.support_)
```
